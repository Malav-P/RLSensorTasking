```python
import sys
sys.path.append("../src")

import numpy as np
import matplotlib.pyplot as plt

from SensorTasking import KalmanFilter, Dynamics
from cr3bp import cr3bp, jac_cr3bp
```

```python
mu = 1.215058560962404e-02
x0 = np.array([1.1540242813087864,
        -2.4170502654082446e-28,
        -0.1384196144071876,
        4.06530060663289e-15,
        -0.21493019200956867,
        8.48098638414804e-15])
P0 = 0.001*np.eye(6)

timestep = 0.015
```

```python
my_filter = KalmanFilter(timestep=timestep, xof=x0, Pof=P0, func=cr3bp, jac=jac_cr3bp, params =(mu,))
truth = Dynamics(x0=x0, tstep=timestep, f=cr3bp, jac=jac_cr3bp, f_params=[mu], jac_params=[mu])
```

```python
def some_function(truth, num_observers = 3):

    dim = truth.shape[0]

    R_invs = np.zeros(shape=(dim, dim, num_observers))
    Z = np.zeros(shape=(dim, num_observers))

    for i in range(3):
        R_invs[i, i, :] = (1/0.01)   # positional uncertaintiy of +- 384 km (VERY CONSERVATIVE, most sensor can do better)
        R_invs[3 + i, 3 + i, :] = (1/0.01) # velocity uncertainty ~ +- 10 m/s

    Z_transpose = np.array([truth.x + np.random.multivariate_normal(mean=np.zeros(dim), cov=np.linalg.inv(R_invs[:,:,j])) for j in range(num_observers)])
    Z = Z_transpose.T

    return Z, R_invs
```

```python
my_filter.reset()
truth.reset()
steps_=400
state_history = np.zeros(shape=(6, steps_))
true_history = np.zeros_like(state_history)

for i in range(steps_):
    truth.propagate(steps=1)
    if steps_%1 == 0:
        Z, R_invs = some_function(truth, num_observers=np.random.randint(1, 5))
    else:
        Z, R_invs = (None, None)
    my_filter._forecast(steps=1)
    my_filter._update(Z, R_invs)

    state_history[:,i] = my_filter.xa
    true_history[:,i] = truth.x
```

```python
plt.scatter(np.arange(0, true_history.shape[1]), true_history[0,:], marker='.', color="black")
plt.scatter(np.arange(0, state_history.shape[1]), state_history[0,:], marker='.', color="red")
```
